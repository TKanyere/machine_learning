Here, we will be looking at an example of the basic principle of monte-carlo sampling, by attempting to estimate pi using a graph. The image below is a summary of the code I wrote.


```python
import random
import matplotlib.pyplot as plt
import numpy as np
def est_pi(n):
#     Set the number of points inside and outside the circle to 0
    num_inside=0
    num_total=0
#     list the co-ordinates of all of the points plotted
    a=[]
    b=[]
#     Visualise the circle curve inside the graph
    theta=np.linspace(0,2*np.pi,100)
    r=1
    x1=r*np.cos(theta)
    x2=r*np.sin(theta)
#     For every point passed through the algorithm, gather the co-ordinates of the point and measure it's 
#     distance from the circle line, determining whether or not the point is inside or outsde of the circle,
#     and record it. All points are uniformly and randomly distributed accross the graph.
    for _ in range(n):
        x=random.uniform(0,1)
        a.append(x)
        y=random.uniform(0,1)
        b.append(y)
        distance=(x**2+y**2)  # multiply co-ordinates of the points.
        if distance<=1:       # Any multipied points inside of the circle will be less than or equal to 1.
            num_inside+=1
        num_total+=1
    func=num_inside/num_total # Ratio of no. of points inside circle to total no. of points.
    print('pi=',4*func)       # Ratio of no. of points inside circle to total no. of points multiplied by 4 = pi.
  
    # Monte Carlo pi estimation visualisation
    plt.figure(figsize = (8, 8))
    plt.ylim(0, 1)
    plt.xlim(0,1)
    plt.scatter(a,b,marker='x',s=0.1)
    plt.title('Estimation of pi using monty hall')
    plt.plot(x1,x2,color='r')
    
est_pi(100000)

```

As you can see, by creating a 1x1 graph, and imposing a quarter circle onto the graph, we end up with having 1 as the radius of the circle, which is the number we are using when calculating pi in the algorithm.
